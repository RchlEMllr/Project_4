```python
 # Import our dependencies
import pandas as pd
from pathlib import Path
```

```python
mt_file_path = Path(".\Resources\murder_target.csv")
crime_df = pd.read_csv(mt_file_path, encoding='cp1252')
crime_df
```

```python
crime_df.isna().sum()
```

```python
crime_na_df = crime_df.dropna()
```

```python
crime_na_df.isna().sum()
```

```python
crime_na_df
```

```python
crime_na_df['precinct_id'].value_counts()
```

```python
crime_na_df = crime_na_df.loc[crime_na_df['precinct_id'] != 999]
crime_na_df = crime_na_df.loc[crime_na_df['precinct_id'] != 316]
crime_na_df = crime_na_df.loc[crime_na_df['precinct_id'] != 232]
crime_na_df = crime_na_df.loc[crime_na_df['precinct_id'] != 532]
crime_na_df['precinct_id'].value_counts()
```

```python
crime_na_df['reported_date'].value_counts()
```

```python
crime_na_df['reported_date'] = pd.to_datetime(crime_na_df['reported_date'])
```

```python
crimes_2019_df = crime_na_df[crime_na_df['reported_date'].dt.year == 2019]
```

```python
crimes_2019_df = crimes_2019_df.reset_index()
crimes_2019_df 
```

```python
crimes_2019_df = crimes_2019_df.drop(columns=['level_0', 'index'])
crimes_2019_df
```

```python
crime_na_df.to_csv('.\Resources\crimes_trimmed.csv', index=False)
```
