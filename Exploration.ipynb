```python
 # Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
from pathlib import Path
```

```python
crimes_file_path = Path(".\Resources\crime.csv")
crime_df = pd.read_csv(crimes_file_path, encoding='cp1252')
crime_df.head()
```

```python
offenses_file_path = Path(".\Resources\offense_codes.csv")
offenses_df = pd.read_csv(offenses_file_path, encoding='cp1252')
offenses_df.head()
```

```python
crime_df = crime_df.drop(columns=['incident_id', 'offense_id', 'first_occurrence_date', 'last_occurrence_date', 'geo_x', 'geo_y', 
                       'is_crime', 'is_traffic'])
crime_df.head()
```

```python
offenses_df = offenses_df.drop(columns=['OBJECTID', 'OFFENSE_CATEGORY_NAME', 'IS_CRIME', 'IS_TRAFFIC'])
offenses_df.head()
```

```python
crime_df.isna().sum()
```

```python
column_names = list(offenses_df.columns.values)
column_names
```

```python
lc_n = [x.lower() for x in column_names]
    
lc_n
```

```python
offenses_df.columns=lc_n
offenses_df.head()
```

```python
offenses_df.dtypes
```

```python
crime_df.dtypes
```

```python
merged_df = pd.merge(crime_df, offenses_df,  how='outer', left_on=['offense_code','offense_code_extension','offense_type_id',
                                                                   'offense_category_id'], right_on=['offense_code',
                                                                                                    'offense_code_extension',
                                                                                                     'offense_type_id',
                                                                                                     'offense_category_id']) 
merged_df.head()
```

```python
merged_df.to_csv('crimes_and_offenses.csv', index=False)
```
