```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as skl
import tensorflow as tf
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
murder_df = pd.read_csv('./Resources/crimes_2019.csv')
murder_df
```

```python

murder_cat = murder_df.dtypes[murder_df.dtypes == "object"].index.tolist()

# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse_output=False)

# Fit and transform the OneHotEncoder using the categorical variable list
code_df = pd.DataFrame(enc.fit_transform(murder_df[murder_cat]))

# Add the encoded variable names to the dataframe
code_df.columns = enc.get_feature_names_out(murder_cat)
```

```python
murder_code_df = murder_df.merge(code_df,left_index=True, right_index=True)
murder_code_df = murder_code_df.drop(murder_cat,1)
murder_code_df.head()
```

```python
y = murder_code_df["murders_1"].values
X = murder_code_df.drop(["murders_1","murders_0"],1).values

# Use sklearn to split dataset


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
